<h5><b>Import Library + Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("AirPassengers.csv")
df.columns = ['Date','Number of Passengers']
df.head(5)
```

<h5><b> Visualize the Data

```python
def plot_df(df, x, y, title="", xlabel='Bulan', ylabel='Jumlah Penumpang', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=45)
    plt.show()
    

plot_df(df, x=df['Date'], y=df['Number of Passengers'], title='Jumlah Penumpang US Airline Tahun 1949-1952')
```

<h5><b>Interpreted Data and Analysis

Data diatas adalah data time-series (runtun waktu) jumlah penumpang US Airline pada tahun 1949-1952.<br>Data tersebut diawali pada bulan Januari 1949 hingga bulan Desember 1952.<br>Dengan melihat grafik dapat diketahui bahwa karakteristik data tersebut termasuk dalam data non-stasioner karena rataan dan ragam bergerak secara tidak konstan.<br>Pola data tersebut adalah pola seasonal, karena memiliki pola musiman berupa kemiripan kenaikan dan penurunan pada bulan tertentu.<br><br>Dari grafik tersebut terlihat kenaikan setiap bulan Februari dan Juni, penurunan setiap bulan Maret hingga April kemudian naik lagi pada bulan Mei.<br>Jumlah wisatawan tertinggi pada bulan Juli 1952 dan paling rendah pada bulan Oktober 1949.<br>Pola kenaikan terjadi karena pada bulan Februari dan Juni di Amerika Serikat sedang dalam liburan musim panas, sehingga orang cenderung melakukan perjalanan jarak jauh di bulan tersebut.<br>Cuaca pada bulan Februari dan Juni cenderung memiliki cuaca yang stabil dan cerah membuatnya lebih menyenangkan untuk melakukan perjalanan.<br>Pada bulan tersebut dikaitkan pada musim pariwisata di daerah-daerah tertentu.Banyak event khusus juga kemungkinan terjadi pada bulan-bulan tersebut.<br>Dan faktor ekonomi juga bisa mempengaruhi kecenderungan kenaikan tersebut karena pada bulan tersebut dimana bonus atau intensif tahunan pekerja biasanya diberikan.<br><br>
